```python
# create function to extract prize list from a given url
def extract_prize_ls(url):
    # start chrome
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument("user-agent=foo")
    #options.add_argument('--headless')
    path="chromedriver.exe"
    driver = webdriver.Chrome(executable_path=path, options=options)
    
    # Access url and find table
    driver.get(url)
    WebDriverWait(driver,1).until(
                            EC.presence_of_element_located((By.XPATH, "//table[@class='table mb-1 mb-md-3']")))
    source = driver.page_source
    soup=bs.BeautifulSoup(source,'html.parser')
    table = soup.find("table",{"class":"table mb-1 mb-md-3"})
    
    # Collecting Ddata
    prizes_ls = []
    for row in table.tbody.find_all('tr'):
        if row.td != None:
            prizes = row.find_all('td')
            for prize in prizes:
                prizes_ls.append(prize.text.split(" "))
                
    # join list of list into 1 list
    prizes_ls = list(itertools.chain.from_iterable(prizes_ls))
    return prizes_ls
```

```python
# create list of all dates in 2014

def create_dates_ls(year: str):
    """
    Input:
    year: in format 'YYYY'
    
    """
    start_date = f"{year}0101"
    end_date = f"{year}1231"
    dates_ls = []
    for dt in rrule.rrule(rrule.DAILY,
                          dtstart=datetime.strptime(start_date, '%Y%m%d'),
                          until=datetime.strptime(end_date, '%Y%m%d')):
        dates_ls.append(dt.strftime('%Y-%m-%d'))
    return dates_ls
```

```python
dates_ls = create_dates_ls('2014')
dates_ls
```

```python
url = "https://www.gidapp.com/lottery/singapore/4d/date/" + dates_ls[0]
print(url)
prizes_ls = extract_prize_ls(url)
prizes_ls
```

```python
failed_dates_ls = []
results_dict = {}   # dict to store prizes and date, will convert into dataframe afterwards
for date in dates_ls:
    url = "https://www.gidapp.com/lottery/singapore/4d/date/" + date
    print(url)
    try:
        prizes_ls = extract_prize_ls(url)
        results_dict[date] = prizes_ls
        print(f"{date} successful")
    except KeyboardInterrupt:
        print('interrupted!')
        break
    except:
        print(f"{date} failed")
        failed_dates_ls.append(date)
```

```python
num_ls = []  # list for numbers column
date_ls = [] # list for dates column
first_prize_ls = [] # list for 1st prize column

for (columnName, columnData) in result_df.iteritems():
    # append prizes for date
    print(columnData.values)
    num_ls.append(columnData.values)
    first_prize_ls.append(columnData.values[0])
    
    # append current date and fill remaining rows with None until next date
    date_append = [columnName,]  # values for current date to append to date column
    for i in range(len(columnData.values)-1):
        date_append.append(None)
    print(date_append)
    date_ls.append(date_append)

# join list of list into 1 list
num_ls = list(itertools.chain.from_iterable(num_ls))
date_ls = list(itertools.chain.from_iterable(date_ls))
print(num_ls, date_ls,first_prize_ls)
```
